```python
dates = []
times = []
sizes = []
prices = []

step_size = 5 #Common difference in the sequence

# Parse the data and store it into lists
for i, size_tag in enumerate(size_tags, start=1):
    if (i - 2) % step_size == 0:
        dates.append(size_tag.text)
    elif (i - 3) % step_size == 0:
        times.append(size_tag.text)
    elif (i - 4) % step_size == 0:
        sizes.append(size_tag.text)
    elif (i - 5) % step_size == 0:
        prices.append(size_tag.text)
    else: pass

for i in range(1, len(prices) + 1):
    print(dates[i], times[i], sizes[i], prices[i])
```

```python
# Create a list of dictionaries to represent the table
table_data = []
for date, time, size, price in zip(dates, times, sizes, prices):
    table_data.append({
        "Date": date,
        "Time": time,
        "Size": size,
        "Price": price
    })

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(table_data)

# Display the DataFrame
print(df)
```
